### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

```python
users.groupby('occupation')['age'].describe()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
gends = users.groupby(['occupation', 'gender'])['gender'].count()
gends
```

```python
gends.index
```

```python
gender_ratio = gends.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum()))
```

```python
gender_ratio
```

```python
gender_ratio.index.get_level_values('gender') == 'M'
```

```python
gender_ratio[gender_ratio.index.get_level_values('gender') == 'M'].sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_by_occu = users.groupby('occupation')['age'].min()
max_by_occu = users.groupby('occupation')['age'].max()
min_max_occu = pd.concat([min_by_occu, max_by_occu], axis=1)
min_max_occu.columns = ['Youngest', 'Oldest']
min_max_occu
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']['age']).mean()
```
